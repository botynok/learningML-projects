Подключение библиотек

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

Запишем исходные данные в DataFrame

```python
df = pd.read_csv("candy-data.csv", delimiter=',', index_col='competitorname')
df      
```

По условию формируем тренировочный набор без некоторых конфет

```python
trainData = df.drop(['One dime', 'Nestle Butterfinger'])
trainData
```

Разделяем выборку на предикторы и отклик, применяем линейную регрессию

```python
X = pd.DataFrame(trainData.drop(['winpercent', 'Y'], axis=1))
y = pd.DataFrame(trainData['winpercent'])

reg = LinearRegression().fit(X, y)

print(*reg.intercept_) #intercept
print(*reg.coef_) #slope

```

Предсказание отклика для некоторых объектов исследования, вывод фактических значений для сравнения. 

```python
kitKat = df.loc['Kit Kat',:].to_frame().T
print('Kit Kat - ', *reg.predict(kitKat.drop(['winpercent', 'Y'], axis=1)), df.loc['Kit Kat', 'winpercent'])

oneDime = df.loc['One dime',:].to_frame().T
print('One dime - ', *reg.predict(oneDime.drop(['winpercent', 'Y'], axis=1)), df.loc['One dime', 'winpercent'])

nestle = df.loc['Nestle Butterfinger',:].to_frame().T
print('Nestle Butterfinger - ', *reg.predict(nestle.drop(['winpercent', 'Y'], axis=1)), df.loc['Nestle Butterfinger', 'winpercent'])
```
